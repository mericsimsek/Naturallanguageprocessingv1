```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
derle=[]
```

```python
for i in range(0,1000): #1000 boyutlku var
        yorum=re.sub("[^a-zA-Z]"," ",yorumlar["Review"][i])
        yorum=yorum.lower()
        yorum=yorum.split()
        ps=PorterStemmer()
        yorum=[ps.stem(kelime) for kelime in yorum if not kelime in set(stopwords.words("english"))]
        yorum = ' '.join(yorum) #yorumu al boşluklarla birleştir
        derle.append(yorum)
```

```python
derle
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000)
X=cv.fit_transform(derle).toarray() # 0 ve 1 e indirdik yani pozitif negatif var yok
y= yorumlar["Rating"][:1000] #for range 1000 e kadar listeye eklettik sonucunda X olarak 1000 boyutu olan listeyi fit ettik
# y 10000 boyutluydu 1000 e idirgedik bu şekil
```

```python
from sklearn.model_selection import train_test_split
X.shape
```

```python
y.shape #şu an böyle ama boyut hatası veriypr   chatgptnin sunduğu bu
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42) 
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)#fit eğitmektir  xtrainden öğren y traine örendikten sonra tahmin et   trainler %80lik kısımdır 
y_pred = gnb.predict(X_test)#şimdi ½20lik kısım tahmin  yani yorumlardaki 1 ve 0ları tahmin et  ytestte bizim gerçek değerlerimiz
#bakalım ne kadar doğru çıkarmış  ilişki içinde confusion matrix kullanacam
cm=confusion_matrix(y_test,y_pred) #birbirleri arasındaki ilişki , bakalım ne kadar iyi tahmin etmiş  tahminler ile gerçekler arasında
print(cm) #68 20  2ye 2 lik matris çıktı ve 95+68 /2  yaptı accuracy başarı oranı yani tahmin oranını bbuldu 
          #30 95  30 +20 /2 yappıp 25 ile hata oranı buldu
```
